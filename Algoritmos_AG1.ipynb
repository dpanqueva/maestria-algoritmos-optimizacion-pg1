```python
#Sucesión_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

# Análisis del algoritmo de Fibonacci
## Coste computacional Fibonacci
Estructura divide y vencerás "ingenua", este algoritmo divide en dos subproblemas más pequeños (Fibonacci(N-1) y Fibonacci(N-2)). Pero lo hace sin coordinación entre ramas. Carece de memoria, siendo incapaz de recordar lo que calculó. Su condena es la superposición de subproblemas, ya que no son independientes, ¿Qué quiere decir esto? Que para calcular F(5) necesitamos F(4) y F(3). Pero cuando vuelve a hallar F(4) tenemos que volver a hallar F(3).

Al no recordar, el árbol de llamadas recursivas se duplica a profundidad. Detonando en el tiempo de ejecución haciendo que crezca de manera exponencial O(o^N) donde o=1618.
Según los calculos de la página https://www.bigocalc.com/ nos dice que el algoritmo pasa el 99,9% recalculando lo mismo. Por ejemplo, si ejecutamos Fibonacci(5), la función Fibonnacci(2) se calcula 3 veces.

El espacio de memoria (O(N)) por la pila de llamadas, genera millones de llamadas en total, no ejecutan todas a la vez.
Python va bajando por la rama izquierda hasta tocar fondo (profundidad) (N < 2) y luego libera memoria a medida que retorna. Por lo tanto, la cantidad máxima de funciones simultáneamente en la memoria RAM en el punto más critico es proporcional a N(La altura de árbol). Esto tiene una repercusión en la optimización, por ejemplo, si ejecutas Fibonacci(2000) el programa colapsa con un error de recursión antes de quedarse sin memoria RAM, superando el límite de la pila del intérprete.

### Árbol de recorrido
                        Fibonacci(5)
                       /            \
              Fibonacci(4)          Fibonacci(3)
             /          \           /          \
      Fibonacci(3)   Fibonacci(2) Fibonacci(2) Fibonacci(1)
      /       \      /       \     /       \
    Fib(2)  Fib(1)     Fib(1)  Fib(0) Fib(1)  Fib(0)
        /    \
    Fib(1) Fib(0)

# Una solución
Con el algoritmo con MEMO, destaca y nos ayuda a que el crecimiento sea de exponencial a **lineal O(N)**, siendo posible el cálculo desde 0 hasta N una sola vez. Haciendo un trabajo constante por primera vez (O(1)) que consiste en la suma de dos números y guardar el resultado en el diccionario.
Las llamadas recursivas, generan un árbol de ejecución y con MEMO se convierte en dos simples consultas al diccionario.

```python
def fibonacci_opt(N: int, memo: dict[int, int] = None) -> int:
    """Calcula el N-ésimo número de Fibonacci usando recursión con memoización.

    La memoización evita recalcular valores ya obtenidos, reduciendo la
    complejidad de O(2^n) a O(n).

    argumentos:
        N: Posición en la secuencia de Fibonacci (entero no negativo).
        memo: Diccionario caché para almacenar resultados intermedios.

    retornos:
        El N-ésimo número de Fibonacci (F(0)=1, F(1)=1).
    """
    if memo is None:
        memo = {}

    if N in memo:
        return memo[N]

    if N < 2:
        resultado = 1
    else:
        resultado = fibonacci_opt(N - 1, memo) + fibonacci_opt(N - 2, memo)

    memo[N] = resultado
    return resultado

fibonacci_opt(5)
```

### Conclusión
El algorimo optimiza tiempo, sacrificando una cantidad aceptable de espacio.
Espacialmente podemos decir que bajamos la complejidad temporal de O(1,618) a O(N), teniendo en cuenta un coste de espacio **O(N)**.

La diferencia entre ambos algoritmos se reduce a una sola palabra **Memoria**
El código original Fibonacci(5) es la representación del esfuerzo repetitivo, en cambio la implementación con memorización representa una eficiencia inteligente en no recordar lo que ya ha recorrido.

## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
